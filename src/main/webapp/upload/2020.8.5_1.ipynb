### 색 약자

 * k : 검정색
 * g : 녹색

 * y : 노란색
 * w : 흰색
### 표준 스타일 약자 
 * p : 오각형 표시
 * o : 원표시
 * . : 점표시
 * s : 사각형 표시
 * - : 실선
 * -. : 대시-점선
 * : : 점선
 * v : 아래쪽 삼각형 표시 
 * * : 별표시
 * d : 다이아몬드 표시

```python
plt.figure(figsize=(15,8))
plt.plot(y.cumsum(), 'b', lw=1.5)
plt.plot(y.cumsum(), 'ks') #검정색 사각형 표시, (색깔 약자 / 점 스타일 조합 )
plt.grid(True) # 그림에 grid 설정
plt.axis('tight') # 축 간격을 조밀하게 설정
plt.xlabel('index') # x축에 이름 붙이기 
plt.ylabel('value') # y축에 이름 붙이기 
plt.title("Example Plot") # 타이틀 설정 
plt.show()
```

### subplots() 메소드의 옵션
 * nrows : 서브플롯의 로우 수 
 * ncols : 서브플롯의 칼럼 수 
 * sharex : 모든 서브플롯이 같은 x축 눈금을 사용
 * sharey : 모든 서브플롯이 같은 y축 눈금을 사용

```python
fig, ax1 = plt.subplots()
plt.plot(y2[:,0], lw=1.5, label='samsung')
plt.plot(y2[:,0], 'ro')
plt.grid(True) # grid 설정
plt.axis('tight') # 축 간격 설정
plt.legend(loc=8)
plt.xlabel('index') # x축  이름 
plt.ylabel('samsung') # y축 이름 
plt.title("Example Plot") # 타이틀 설정 
ax2 = ax1.twinx()                         # twinx 명령은 x 축을 공유하는 새로운 ax2 객체를 생성_ 축을 다르게 하기 위해서
plt.plot(y2[:,1], 'g', lw=1.5, label='namyang')
plt.plot(y2[:,1], 'ro')
plt.legend(loc=0)
plt.ylabel('namyang ')
```

### 여러 그래프를 그리는 방법 
* 1. subplot(row, col, index)
* 2. add_subplot()
* 3. subplots()

```python
plt.figure(figsize=(15,8))
plt.subplot(2,1,1) #row, col, index
plt.plot(y2[:,0], lw=1.5, label='samsung')
plt.plot(y2[:,0], 'ro')
plt.grid(True) 
plt.axis('tight')
plt.legend(loc=0)
plt.xlabel('index')
plt.ylabel('samsung')
plt.title("Example Plot")
plt.subplot(2,1,2)
plt.plot(y2[:,1], 'g', lw=1.5, label='namyang')
plt.plot(y2[:,1], 'ro')
plt.grid(True) 
plt.axis('tight') 
plt.legend(loc=0)
plt.ylabel('namyang ')
plt.show()
```

```python
fig = plt.figure(figsize=(15,8))

ax1  = fig.add_subplot(2,1,1)
ax1.plot(y2[:,0], lw=1.5, label='samsung')
ax1.plot(y2[:,0], 'ro')
ax1.grid(True)
ax1.axis('tight')
ax1.legend(loc=0)
ax1.set_xlabel('index')    # 객체를 호출해서 만들때는 set_xlabel, set_title 등 처럼 이렇게 만든다. 원래는 plt.xlabel, plt.title
ax1.set_ylabel('samsung')
ax1.set_title("Example Plot")
ax2 = fig.add_subplot(2,1,2)
ax2.plot(y2[:,1], 'g', lw=1.5, label='namyang')
ax2.plot(y2[:,1], 'ro')
ax2.grid(True) 
ax2.axis('tight') 
ax2.legend(loc=0)
```

```python
fig, ax = plt.subplots(2,1, figsize=(15,8))

ax[0].plot(y2[:,0], lw=1.5, label='samsung')
ax[0].plot(y2[:,0], 'ro')

ax[1].plot(y2[:,1], 'g', lw=1.5, label='namyang')
ax[1].plot(y2[:,1], 'ro')

plt.show()
```
